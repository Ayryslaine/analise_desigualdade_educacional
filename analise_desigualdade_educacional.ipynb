# Tratamento e limpeza dos dados

Para começar, vamos verificar os tipos de dados que temos em nosso dataframe, quantidade de colunas e quantidade de dados

```python
df.info()
```

  Em nosso banco de dados, temos o total de 12 colunas, com 1.000 entrada de dados e com seus dados divididos em object, int e float

Tendo as informações gerais sobre o banco de dados, agora vamos verificar se existem valores nulos ou registros duplicados.

```python
verificar_nulos = df.isnull().sum()
display(verificar_nulos)
```

```python
verificar_duplicados = df.duplicated().sum()
display(verificar_duplicados)
```

Podemos visualizar que em nosso banco de dados não existe valores nulos/faltantes ou duplicados

Agora vamos renomear as colunas para melhor entendimento

```python
df.rename(columns= {
    "school_name":"Nome da escola",
    "state":"Estado",
    "school_type":"Tipo de instituição",
    "grade_level":"Nível escolar",
    "funding_per_student_usd":"Financiamento por aluno US",
    "avg_test_score_percent":"Desempenho do aluno",
    "student_teacher_ratio":"Alunos por professor",
    "percent_low_income":"Estudantes de baixa renda",
    "percent_minority":"Origens minoritárias",
    "internet_access_percent":"Acesso à internet (%)",
    "dropout_rate_percent":"Evasão Escolar (%)"
}, inplace=True)

display(df)
```

Agora que modificamos os nomes das colunas, vamos remover algumas colunas que não serão necessárias em nossa análise, sendo elas: a coluna id e a coluna Nome da escola

```python
df.drop(columns={
    "id",
    "Nome da escola"
}, inplace=True)

display(df)
```
